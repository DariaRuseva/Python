# Creates a map visualising the top 10 most used - in orange - (least used- in red) starting stations for electric bike users

```python
import folium 

top_start_stations = electric_bikes['start_station_name'].value_counts().head(10)
least_used_start_stations = electric_bikes['start_station_name'].value_counts().tail(10)

map_center = [electric_bikes['start_lat'].mean(), electric_bikes['start_lng'].mean()]
map_dc = folium.Map(location=[38.8951100, -77.0363700], zoom_start=14, zoom_end = 26)

#bounding box 
bbox = [38.75, -77.25, 39.05, -76.80]  
folium.Rectangle(bounds=[(bbox[0], bbox[1]), (bbox[2], bbox[3])], color='red',
                 fill=False, weight=2, opacity=1).add_to(map_dc)

# Adds  GeoJSON layer
folium.GeoJson("bike_paths_dc.geojson", name="Bike Paths").add_to(map_dc)

# orange for the top 10 start stations
for i, (station_name, count) in enumerate(top_start_stations.items()):
    station_data = electric_bikes[electric_bikes['start_station_name'] == station_name].iloc[0]
    
    # blue for the first top station
    marker_color = 'blue' if i == 0 else 'orange'

    folium.Marker(
        location=[station_data['start_lat'], station_data['start_lng']],
        popup=f"{station_name}: {count} Rides",
        icon=folium.Icon(color=marker_color, icon='bolt', prefix='fa')
    ).add_to(map_dc)

#adds markers for the least used 10 start stations
for station_name, count in least_used_start_stations.items():
    station_data = electric_bikes[electric_bikes['start_station_name'] == station_name].iloc[0]
    
    # red for the least used stations
    marker_color = 'red'
    
    # lightning bolt icon for all markers
    folium.Marker(
        location=[station_data['start_lat'], station_data['start_lng']],
        popup=f"Least Used Station: {station_name}: {count} rides",
        icon=folium.Icon(color=marker_color, icon='bolt', prefix='fa')
    ).add_to(map_dc)
# Saves the map as an HTML file
map_dc.save("top_start_stations_map.html")
```

# Some KPIs ( Average Distance per Electric Bike Trip by Year, Electric Bike Adoption Rate Growth) 

```python
import pandas as pd
import geopy.distance  

electric_bike_trips = bikes[bikes['rideable_type'] == 'electric_bike']

def calculate_distance(row):
    start_coords = (row['start_lat'], row['start_lng'])
    end_coords = (row['end_lat'], row['end_lng'])
    return geopy.distance.distance(start_coords, end_coords).km

electric_bike_trips['distance'] = electric_bike_trips.apply(calculate_distance, axis=1)

# Extract year from 'started_at' column
electric_bike_trips['year'] = electric_bike_trips['started_at'].dt.year

average_distance_per_trip_by_year = electric_bike_trips.groupby('year')['distance'].mean()

print("Average Distance per Electric Bike Trip by Year:")
print(average_distance_per_trip_by_year)
```

```python
# Calculate the number of electric bike rides for each year
electric_bike_rides_by_year = electric_bike_trips.groupby('year').size()

# Calculate Electric Bike Adoption Rate Growth for each year compared to the previous year
growth_2022 = ((electric_bike_rides_by_year[2022] - electric_bike_rides_by_year[2021]) / electric_bike_rides_by_year[2021]) * 100
growth_2023 = ((electric_bike_rides_by_year[2023] - electric_bike_rides_by_year[2022]) / electric_bike_rides_by_year[2022]) * 100

print("Electric Bike Adoption Rate Growth:")
print(f"Growth from 2021 to 2022: {growth_2022:.2f}%")
print(f"Growth from 2022 to 2023: {growth_2023:.2f}%")
```

# Performs Supervised Linear and Polynomial Regression 

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, PolynomialFeatures
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LinearRegression


features = ['temperature_2m (°C)', 'relativehumidity_2m (%)', 'started_at']

# Target column
target = merged_data["Total_rides"] 

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(merged_data[features], merged_data[target], test_size=0.2, random_state=42)

# Separate transformers for numerical and datetime features
numeric_features = ['temperature_2m (°C)', 'relativehumidity_2m (%)']
datetime_features = ['started_at']

# Use ColumnTransformer to apply different transformers to different columns
preprocessor = ColumnTransformer(
    transformers=[
        ('numeric', PolynomialFeatures(degree=2, include_bias=False), numeric_features),
        ('datetime', Pipeline(steps=[
            ('onehot', OneHotEncoder(handle_unknown='ignore')),
            ('poly', PolynomialFeatures(degree=2, include_bias=False))
        ]), datetime_features)
    ])

# Linear Regression
linear_model = LinearRegression()
linear_pipeline = Pipeline(steps=[('preprocessor', preprocessor), ('regressor', linear_model)])
linear_pipeline.fit(X_train, y_train)
y_pred_linear = linear_pipeline.predict(X_test)

# Polynomial Regression
poly_model = LinearRegression()
poly_pipeline = Pipeline(steps=[('preprocessor', preprocessor), ('regressor', poly_model)])
poly_pipeline.fit(X_train, y_train)
y_pred_poly = poly_pipeline.predict(X_test)

```
