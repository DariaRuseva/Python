Let's now group our qualities into 3 categories: low, medium and high quality wines.

```python
# we are creating a new column called "quality_label", we define a range and associate that range with a label
red_wine['quality_label'] = red_wine['quality'].apply(lambda value: 'low'
if value <= 5 else 'medium'
if value <= 7 else 'high')
#lambda  keyword is used to define an anonymous (unnamed) 
# function.value: This is the parameter of the lambda function. When you apply this 
# lambda function using red_wine['quality'].apply(...), it will be passed each value in the 'quality' column of the DataFrame.

```

```python
#Doing the same for the white wine
white_wine["quality_label"]=white_wine["quality"].apply(lambda value: "low" 
if value <=5 else "medium"
if value <=7 else "high")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


#fig, ax = plt.subplots(white_wine["quality_label"], red_wine["quality_label"]) # wrong because i need a singe chart
#ax.pie(sizes, labels=labels) 
#red_wine is an integer; no need for len function

# Data for the pie chart
labels = ['Red Wines', 'White Wines']
sizes = [len(red_wine),len(white_wine)]
colors = ['#ff9999', '#66b3ff']

# Create a pie chart
fig, ax = plt.subplots()
ax.pie(sizes, labels=labels,autopct='%1.1f%%', startangle=90, colors=colors)

# Draw a circle at the center to make it look like a donut chart
#centre_circle = plt.Circle((0, 0), 0.70, fc='white')
#fig = plt.gcf()
#fig.gca().add_artist(centre_circle)

# Equal aspect ratio ensures that pie is drawn as a circle
ax.axis('equal')  

# Display the pie chart
plt.title('Quantity of Wines')
plt.show()
```

```python

# Show the plot
pd.set_option('mode.use_inf_as_na', False)

plt.title('Distribution of Red Wine Qualities')
plt.xlabel('Quality Label')
plt.ylabel('Count')
quality_order = ['low', 'medium', 'high']#!!! important for comparisson


sns.histplot(data=len(red_wine), x=red_wine["quality_label"], color='#ff9999',edgecolor='grey', linewidth=0.2)
plt.ylim(0, 3000) #ensures that both hist are comparble, aka have the same Y-axis 
red_hist = sns.histplot(data=red_wine, x="quality_label", color='#ff9999', edgecolor='grey', linewidth=0.2)
red_hist.set_xticklabels(quality_order)
plt.show()

```

```python
plt.title('Distribution of White Wine Qualities')
plt.xlabel('Quality Label')
plt.ylabel('Count')
quality_order = ['low', 'medium', 'high']

white_hist=sns.histplot(data=len(white_wine), x=white_wine["quality_label"], color='#66b3ff',edgecolor='grey', linewidth=0.5) #to set the order of the labels
white_hist.set_xticklabels(quality_order)
plt.show()

```

```python
#Now try to compare the quality with some other features
# Concatenate the two DataFrames into a single DataFrame
combined_df = pd.concat([red_wine, white_wine], ignore_index=True)
my_palette = sns.color_palette("Set2") #sets a pastel palette
# Create a boxplot with hue (red and white wines on the same axis)
plt.figure(figsize=(6, 4))  # Adjust the figure size as needed
boxplot = sns.boxplot(x='quality_label', y='alcohol', data=combined_df, palette=my_palette)

plt.xlabel('Wine Quality Class')
plt.ylabel('Alcohol %')
plt.xticks(rotation=45)


quality_order = ['low', 'medium', 'high']
boxplot.set_xticklabels(quality_order)

plt.title('Wine Quality-Alcohol Content/pH')
plt.show()
```

```python
my_palette = sns.color_palette("Set2") 
sns.boxplot(x=white_wine["quality_label"],y=white_wine["alcohol"],palette=my_palette) 
#x and y-axis labels
plt.xlabel('Wine Quaity')
plt.ylabel('Alcohol %')
plt.xticks(rotation=45)
quality_order = ['low', 'medium', 'high']
white=sns.boxplot(x=white_wine["quality_label"], y=white_wine["alcohol"])
white.set_xticklabels(quality_order)
#pllot title
plt.title('White wine Quality-Alcohol Content/pH')
```

```python
y_palette = sns.color_palette("Set2") 
boxplot = sns.boxplot(x='quality_label', y='pH', data=combined_df, palette=my_palette)

plt.xlabel('Wine Quality Class')
plt.ylabel('Wine pH')
plt.xticks(rotation=45)

quality_order = ['low', 'medium', 'high']
boxplot.set_xticklabels(quality_order)

plt.title('Wine-pH')
plt.show()
```

```python
selected_columns = ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 'alcohol', 'quality']

# Create a pairplot
pairplot = sns.pairplot(combined_df[selected_columns])
plt.suptitle('Pairplot of Selected Variables by Wine Type', y=1.02)
plt.show()
```
